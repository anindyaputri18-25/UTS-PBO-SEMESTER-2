```python
from datetime import datetime  # Mengambil library untuk mencatat waktu aktivitas

class PharmaStock:
    def __init__(self, nama_admin, email, role):
        # --- DATA PROFIL (Private) ---
        self.__admin = nama_admin      # Menyimpan nama admin secara rahasia
        self.__email = email           # Menyimpan email admin secara rahasia
        self.__role = role             # Menyimpan peran/jabatan user (Kasir/Admin/Manager Gudang/Apoteker)

        # --- DATABASE STOK OBAT (Private) ---
        self.__obat = {
            "Paracetamol": {"stok": 120, "exp": "2029-12-01", "supplier": "PT. Kimia Farma", "harga": 5000},
            "Panadol": {"stok": 256, "exp": "2090-08-15", "supplier": "PT. Sterling", "harga": 8500},
            "Amoxicillin": {"stok": 85, "exp": "2089-01-20", "supplier": "PT. Sanbe Farma", "harga": 12000},
            "Promag": {"stok": 50, "exp": "2046-05-10", "supplier": "PT. Kalbe Farma", "harga": 3000},
            "Sanmol": {"stok": 100, "exp": "2099-11-30", "supplier": "PT. Caprifarmindo", "harga": 5500}
        }

        self.__log_aktivitas = []      # List kosong untuk menampung histori aksi user

    def lihat_profil(self):
        print("\n" + "="*40)
        print(f"{' AKUN DASHBOARD PHARMA ':^40}") # Judul profil rata tengah
        print("="*40)
        print(f"👤 Nama Pengguna : {self.__admin}")
        print(f"📩 Email Aktif   : {self.__email}")
        print(f"🔑 Role Akses    : {self.__role}")
        print("-" * 40)

    def update_stok_manual(self, nama_obat, jumlah_baru, tgl_exp = None, supplier = None):
        """Fitur untuk input stok obat secara manual oleh admin atau manager gudang"""
        print(f"\n[SISTEM] Mencoba memperbarui data: {nama_obat}...")

        if jumlah_baru < 0: # Validasi: Stok tidak boleh minus
            self.__catat_log(f"Gagal update {nama_obat}: Jumlah negatif") # Catat kegagalan di log
            print(f"❌ ERROR : Stok {nama_obat} tidak boleh di bawah 0")
            return

        if nama_obat in self.__obat: # Cek apakah obat ada di database
            stok_lama = self.__obat[nama_obat]["stok"] # Ambil stok lama untuk laporan log
            self.__obat[nama_obat]["stok"] = jumlah_baru # Timpa stok lama dengan stok baru

            if tgl_exp: # Jika tgl_exp diisi, update datanya
                self.__obat[nama_obat]["exp"] = tgl_exp
            if supplier: # Jika supplier diisi, update datanya
                self.__obat[nama_obat]["supplier"] = supplier

            # Catat keberhasilan ke log aktivitas
            msg = f"Update Stok {nama_obat} dari {stok_lama} ke {jumlah_baru}"
            self.__catat_log(msg)
            print(f"✅ Berhasil : Data {nama_obat} telah diperbarui di database")
        else:
            print(f"❌ Gagal : Data {nama_obat} tidak ditemukan di database")

    def jual_obat(self, nama_obat, jumlah_jual):
        """Fitur untuk mengurangi stok saat ada transaksi penjualan"""
        if nama_obat in self.__obat:
            stok_obat = self.__obat[nama_obat]["stok"] # Cek stok yang ada sekarang
            if stok_obat >= jumlah_jual: # Validasi: Stok cukup atau tidak
                self.__obat[nama_obat]["stok"] -= jumlah_jual # Kurangi stok sesuai jumlah jual
                total_harga = jumlah_jual * self.__obat[nama_obat]["harga"] # Hitung total uang
                self.__catat_log(f"Jual {nama_obat} sebanyak {jumlah_jual} unit") # Masuk ke histori
                print(f"✅ Transaksi Berhasil : {jumlah_jual} {nama_obat} terjual (Total: Rp{total_harga})")
            else:
                print(f"❌ Gagal : Stok {nama_obat} tidak mencukupi (Sisa : {stok_obat})")
        else:
            print(f"❌ Gagal : Data {nama_obat} tidak ditemukan di database")

    def tampilkan_dashboard_stok(self):
        """Menampilkan tabel monitoring stok secara keseluruhan"""
        print("\n" + "—"*85)
        print(f"{'DASHBOARD MONITORING STOK OBAT APOTEK':^85}")
        print("—"*85)
        print(f"{'NAMA OBAT':<15} | {'STOK':<7} | {'HARGA':<10} | {'EXP DATE':<12} | {'SUPPLIER'}")
        print("-" * 85)

        for nama, data in self.__obat.items():
            # Jika stok < 20, tambahkan tanda peringatan ⚠️ (Fitur cerdas dashboard)
            status_stok = f"{data['stok']}⚠️" if data['stok'] < 20 else data['stok']

            # Print baris tabel dengan format rapi (rata kiri)
            print(f"{nama:<15} | {status_stok:<7} | Rp{data['harga']:<8} | {data['exp']:<12} | {data['supplier']}")
        print("—"*85)

    def __catat_log(self, pesan):
        """Private Method untuk mencatat setiap langkah user ke dalam log"""
        waktu = datetime.now().strftime("%Y-%m-%d %H:%M:%S") # Ambil waktu sekarang
        entri_log = f"[{waktu}] {self.__admin} ({self.__role}): {pesan}" # Format pesan log
        self.__log_aktivitas.append(entri_log) # Simpan ke list log

    def tampilkan_log_aktivitas(self):
        """Menampilkan histori laporan untuk audit/laporan harian"""
        print("\n" + "📋 LOG AKTIVITAS SISTEM (HISTORI)")
        print("-" * 60)
        if not self.__log_aktivitas: # Cek jika belum ada aktivitas
            print("Belum ada aktivitas tercatat.")
        for log in self.__log_aktivitas: # Loop dan cetak semua histori
            print(log)
        print("-" * 60)

# 1. Login sebagai Manager Gudang
sistem_apotek = PharmaStock("Susi Antika", "susi@apotek.com", "Manager Gudang")

# 2. Lihat Profil Admin
sistem_apotek.lihat_profil()

# 3. Lihat Dashboard Awal (Isi 5 obat)
sistem_apotek.tampilkan_dashboard_stok()

# 4. Aksi: Update Stok Manual
sistem_apotek.update_stok_manual("Amoxicillin", 200, "2027-05-20", "PT. Sanbe Farma (Pusat)")

# 5. Aksi: Terjadi Penjualan
sistem_apotek.jual_obat("Paracetamol", 10)
sistem_apotek.jual_obat("Promag", 40) # Stok akan sisa 10 (muncul tanda ⚠️ di dashboard)

# 6. Aksi: Update Salah (Negatif)
sistem_apotek.update_stok_manual("Sanmol", -5)

# 7. Tampilkan Dashboard Terbaru setelah aktivitas
sistem_apotek.tampilkan_dashboard_stok()

# 8. Tampilkan Histori Aktivitas (Fitur Laporan)
sistem_apotek.tampilkan_log_aktivitas()
```

inheritance

```python
============================================================

# 1. Class Utama (Grandparent)
class ProdukFarmasi:
    def __init__(self, nama, harga):
        self.nama = nama                # Inisialisasi nama produk dasar
        self.harga = harga              # Inisialisasi harga produk dasar

# 2. Class Anak (Parent untuk Racikan) - Mewarisi ProdukFarmasi
class ObatSatuan(ProdukFarmasi):
    def __init__(self, nama, harga, stok, exp_date, supplier):
        # Mengambil atribut nama dan harga dari ProdukFarmasi menggunakan super()
        super().__init__(nama, harga)
        self.stok = stok                # Menambah atribut stok (Fitur stok manual)
        self.exp_date = exp_date        # Menambah atribut tanggal kadaluarsa
        self.supplier = supplier        # Menambah atribut supplier (Laporan supplier)

    def tampilkan_stok(self):
        # Mencetak baris data obat untuk tampilan Dashboard tabel
        print(f"| {self.nama:<15} | {self.stok:<5} | {self.exp_date:<12} | {self.supplier}")

    def update_stok_manual(self, jumlah_baru):
        # Method untuk mengubah angka stok secara manual
        self.stok = jumlah_baru
        print(f"[SISTEM] Stok {self.nama} berhasil diupdate manual menjadi: {self.stok}")

# 3. Class Cucu (Child dari ObatSatuan) - Mewarisi ObatSatuan
class ObatRacikan(ObatSatuan):
    def __init__(self, nama, harga, stok, exp_date, supplier, list_komposisi, jenis_racikan):
        # Mengambil semua atribut dari ObatSatuan (stok, exp, supplier)
        super().__init__(nama, harga, stok, exp_date, supplier)
        self.list_komposisi = list_komposisi # Menambah data komposisi bahan racikan
        self.jenis_racikan = jenis_racikan   # Menambah kategori (Puyer/Sirup/dll)

    # Fitur Laporan Khusus Racikan
    def tampilkan_detail_racikan(self):
        # Mencetak laporan lengkap hasil gabungan data induk dan data anak
        print("\n" + "="*45)
        print(f"📑 LAPORAN DATA RACIKAN: {self.nama.upper()}")
        print("="*45)
        print(f"Jenis Racikan : {self.jenis_racikan}")
        print(f"Harga Jual    : Rp.{self.harga}")
        print(f"Sisa Stok     : {self.stok} unit")
        print(f"Kadaluarsa    : {self.exp_date}")
        print(f"Supplier Bahan: {self.supplier}")
        print(f"Komposisi     : {', '.join(self.list_komposisi)}")
        print("="*45)

# ============================================================
# IMPLEMENTASI PADA SISTEM (OUTPUT)
# ============================================================

# --- 1. Inisialisasi Data Obat Satuan ---
# Membuat 5 objek awal untuk mengisi Dashboard
obat1 = ObatSatuan("Paracetamol", 5000, 100, "2025-10-10", "PT. Kimia Farma")
obat2 = ObatSatuan("Amoxicillin", 12000, 50, "2024-12-20", "PT. Sanbe Farma")
obat3 = ObatSatuan("CTM", 2000, 200, "2026-01-15", "PT. Medika")
obat4 = ObatSatuan("Promag", 8000, 30, "2025-05-05", "PT. Kalbe")
obat5 = ObatSatuan("Vitamin C", 5000, 150, "2025-08-20", "PT. Kimia Farma")

# --- 2. Dashboard Monitoring Stok ---
print("\n--- DASHBOARD MONITORING STOK OBAT ---")
print("-" * 55)
print(f"| {'NAMA OBAT':<15} | {'STOK':<5} | {'EXP':<12} | {'SUPPLIER'}")
print("-" * 55)
# Memanggil method tampilkan_stok dari masing-masing objek
obat1.tampilkan_stok()
obat2.tampilkan_stok()
obat3.tampilkan_stok()
obat4.tampilkan_stok()
obat5.tampilkan_stok()
print("-" * 55)

# --- 3. Update Stok Manual ---
# Simulasi admin mengubah stok Promag secara manual
obat4.update_stok_manual(45)

# --- 4. Membuat Data Racikan ---
# Objek racikan mewarisi kemampuan dari ObatSatuan dan ProdukFarmasi
racikan_flu = ObatRacikan(
    "Flu Spesial Dewasa",
    25000,
    10,
    "2024-11-01",
    "Internal Apotek",
    ["Paracetamol", "CTM", "Amoxicillin"],
    "Puyer"
)

# --- 5. Laporan Jual Beli / Detail Racikan ---
# Menampilkan output laporan akhir fitur racikan
racikan_flu.tampilkan_detail_racikan()
```

polimorfisme

```python
from datetime import datetime  # Mengambil library waktu untuk mencatat kapan laporan dibuat

# ============================================================
# METODE: POLIMORFISME (Banyak Bentuk)
# ============================================================

# Class Induk sebagai Kontrak/Interface
class DokumenFarmasi:
    def __init__(self):
        # Otomatis mencatat waktu saat objek laporan apa pun dibuat
        self.waktu = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    def cetak_laporan(self):
        """Method ini akan di-override (ditimpa) oleh setiap class anak"""
        pass

# 1. Fitur: Laporan Jual-Beli ke Supplier (Restock)
class LaporanRestock(DokumenFarmasi):
    def __init__(self, supplier, item, jumlah, biaya):
        super().__init__()            # Memanggil waktu dari class induk
        self.supplier = supplier      # Menyimpan nama supplier
        self.item = item              # Nama obat yang dibeli
        self.jumlah = jumlah          # Jumlah yang masuk
        self.biaya = biaya            # Total biaya restock

    def cetak_laporan(self):
        # Polimorfisme: Implementasi cetak khusus untuk restock supplier
        return (f"[{self.waktu}] 📥 LAPORAN RESTOCK (SUPPLIER)\n"
                f"    - Supplier : {self.supplier}\n"
                f"    - Barang   : {self.item} ({self.jumlah} unit)\n"
                f"    - Total Tagihan : Rp.{self.biaya:,}\n")

# 2. Fitur: Laporan Penjualan (Dashboard Aktivitas Kasir)
class LaporanPenjualan(DokumenFarmasi):
    def __init__(self, kasir, item, total_harga):
        super().__init__()            # Memanggil waktu dari class induk
        self.kasir = kasir            # Nama petugas kasir
        self.item = item              # Nama obat terjual
        self.total = total_harga      # Total uang yang diterima

    def cetak_laporan(self):
        # Polimorfisme: Implementasi cetak khusus untuk penjualan harian
        return (f"[{self.waktu}] 📤 LAPORAN PENJUALAN (KASIR)\n"
                f"    - Petugas  : {self.kasir}\n"
                f"    - Item     : {self.item}\n"
                f"    - Pendapatan: Rp.{self.total:,}\n")

# 3. Fitur: Laporan Stok Manual (Perubahan Data oleh Admin)
class LaporanUpdateStok(DokumenFarmasi):
    def __init__(self, admin, obat, stok_lama, stok_baru):
        super().__init__()            # Memanggil waktu dari class induk
        self.admin = admin            # Nama admin gudang
        self.obat = obat              # Nama obat yang diupdate
        self.lama = stok_lama         # Stok sebelum perubahan
        self.baru = stok_baru         # Stok sesudah perubahan

    def cetak_laporan(self):
        # Polimorfisme: Implementasi cetak khusus untuk update manual di dashboard
        return (f"[{self.waktu}] 📝 LAPORAN UPDATE STOK MANUAL\n"
                f"    - Admin    : {self.admin}\n"
                f"    - Obat     : {self.obat}\n"
                f"    - Perubahan: {self.lama} -> {self.baru} unit\n")

# 4. Fitur: Laporan Data Racikan (Komposisi Apoteker)
class LaporanRacikan(DokumenFarmasi):
    def __init__(self, nama_racikan, komposisi, exp_date):
        super().__init__()            # Memanggil waktu dari class induk
        self.racikan = nama_racikan   # Nama racikan obat
        self.komposisi = komposisi    # Daftar bahan komposisi (list)
        self.exp = exp_date           # Tanggal kadaluarsa

    def cetak_laporan(self):
        # Polimorfisme: Implementasi cetak khusus untuk data racikan
        bahan = ", ".join(self.komposisi) # Menggabungkan list komposisi jadi teks
        return (f"[{self.waktu}] 🧪 LAPORAN PRODUKSI RACIKAN\n"
                f"    - Nama Racikan : {self.racikan}\n"
                f"    - Bahan Utama  : {bahan}\n"
                f"    - Kadaluarsa   : {self.exp}\n")

# ============================================================
# IMPLEMENTASI POLIMORFISME PADA DASHBOARD
# ============================================================

# Menampung berbagai jenis objek laporan dalam satu List
# Walaupun bentuknya beda-beda (Restock, Jual, Update, Racikan)
dashboard_aktivitas = [
    LaporanRestock("PT. Kimia Farma", "Paracetamol", 500, 2500000),
    LaporanPenjualan("Susi Antika", "Panadol Extra", 15000),
    LaporanUpdateStok("Admin_Gudang", "Amoxicillin", 50, 150),
    LaporanRacikan("Puyer Flu Anak", ["Paracetamol", "CTM"], "2025-01-01"),
    LaporanRestock("Supplier Medika", "Vitamin C", 200, 1000000)
]

print("="*60)
print(f"{'DASHBOARD RIWAYAT AKTIVITAS APOTEK':^60}")
print("="*60)

# INTI POLIMORFISME:
# Kita tidak perlu repot mengecek jenis laporan satu per satu (if-else).
# Kita cukup memanggil method yang sama (.cetak_laporan) dan Python
# otomatis menjalankan fungsi yang sesuai dengan jenis objeknya.
for item in dashboard_aktivitas:
    print(item.cetak_laporan())
    print("-" * 60)
```
